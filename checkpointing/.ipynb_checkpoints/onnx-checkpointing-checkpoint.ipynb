```python
import torch.onnx
import torch.nn as nn
import torchvision.transforms as transforms
import torchvision.datasets as dsets
import hashlib
import numpy as np
import onnx
import onnxruntime as ort
```

```python
'''
STEP 1: LOADING DATASET
'''

train_dataset = dsets.MNIST(root='./data', 
                            train=True, 
                            transform=transforms.ToTensor(),
                            download=True)

test_dataset = dsets.MNIST(root='./data', 
                           train=False, 
                           transform=transforms.ToTensor())


'''
STEP 2: MAKING DATASET ITERABLE
'''

batch_size = 100
n_iters = 1000
num_epochs = n_iters / (len(train_dataset) / batch_size)
num_epochs = int(num_epochs)

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, 
                                           batch_size=batch_size, 
                                           shuffle=True)

test_loader = torch.utils.data.DataLoader(dataset=test_dataset, 
                                          batch_size=batch_size, 
                                          shuffle=False)
```

```python

'''
STEP 3: CREATE MODEL CLASS
'''        
class FeedforwardNeuralNetModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(FeedforwardNeuralNetModel, self).__init__()
        # Linear function 1: 784 --> 100
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        # Non-linearity 1
        self.relu1 = nn.ReLU()

        # Linear function 2: 100 --> 100
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        # Non-linearity 2
        self.relu2 = nn.ReLU()

        # Linear function 3: 100 --> 100
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        # Non-linearity 3
        self.relu3 = nn.ReLU()

        # Linear function 4 (readout): 100 --> 10
        self.fc4 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        # Linear function 1
        out = self.fc1(x)
        # Non-linearity 1
        out = self.relu1(out)
        # Linear function 2
        out = self.fc2(out)
        # Non-linearity 2
        out = self.relu2(out)

        # Linear function 2
        out = self.fc3(out)
        # Non-linearity 2
        out = self.relu3(out)

        # Linear function 4 (readout)
        out = self.fc4(out)
        return out
'''
STEP 4: INSTANTIATE MODEL CLASS
'''
input_dim = 28*28
hidden_dim = 100
output_dim = 10

model = FeedforwardNeuralNetModel(input_dim, hidden_dim, output_dim)

'''
STEP 5: INSTANTIATE LOSS CLASS
'''
criterion = nn.CrossEntropyLoss()


'''
STEP 6: INSTANTIATE OPTIMIZER CLASS
'''
learning_rate = 0.1

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

'''
STEP 7: TRAIN THE MODEL
'''
iter = 0
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        # Load images with gradient accumulation capabilities
        images = images.view(-1, 28*28).requires_grad_()

        # Clear gradients w.r.t. parameters
        optimizer.zero_grad()

        # Forward pass to get output/logits
        outputs = model(images)

        # Calculate Loss: softmax --> cross entropy loss
        loss = criterion(outputs, labels)

        # Getting gradients w.r.t. parameters
        loss.backward()

        # Updating parameters
        optimizer.step()

        iter += 1

        if iter % 500 == 0:
            # Calculate Accuracy         
            correct = 0
            total = 0
            # Iterate through test dataset
            for images, labels in test_loader:
                # Load images with gradient accumulation capabilities
                images = images.view(-1, 28*28).requires_grad_()

                # Forward pass only to get logits/output
                outputs = model(images)

                # Get predictions from the maximum value
                _, predicted = torch.max(outputs.data, 1)

                # Total number of labels
                total += labels.size(0)

                # Total correct predictions
                correct += (predicted == labels).sum()

            accuracy = 100 * correct / total

            # Print Loss
            print('Iteration: {}. Loss: {}. Accuracy: {}'.format(iter, loss.item(), accuracy))
```

```python
'''
High Level Functions
'''

class Checkpoint:
    '''
    A Checkpoint is a snapshot of a state transition in an onnx graph.
    In general a state transition is the "flow" over some partition of a computation graph.
    '''
    def __init__(inputs, outputs):
        self.inputs = inputs
        self.outputs = outputs

def get_output_info(model, node_name):
    for value_info in model.graph.value_info:
        if value_info.name == node_name:
            return value_info.type
    # If not found in value_info, it might be in the outputs
    for output in model.graph.output:
        if output.name == node_name:
            return output.type
    return None


def instrument_node(node, model, idx, new_nodes):
    # Create an identity node to capture the output of the original node
    instrumented_output_name = f"{node.output[0]}_instrumented_{idx}"
    identity_node = onnx.helper.make_node(
        'Identity',
        inputs=[node.output[0]],
        outputs=[node.output[0], instrumented_output_name],
        name=f"Identity_{idx}"
    )

    # Append the identity node to the new nodes list
    new_nodes.append(identity_node)

    # Append the instrumented output to the value_info
    output_type = get_output_info(model, node.output[0])
    if output_type is not None:
        model.graph.value_info.append(
            onnx.helper.make_value_info(instrumented_output_name, output_type)
        )

    # Return the original node (unchanged)
    return node

def instrument_graph(model):
    '''
    Instruments an onnx model by adding outputs for all intermediate nodes, without changing the function of the graph.
    
                Parameters:
                        model (ONNXModel): The onnx model to instrument
    
                Returns:
                        instrumented_onnx_graph (ONNXGraph): The instrumented onnx graph
    '''

    # Manually copy the graph components
    inputs = [input for input in original_model.graph.input]
    outputs = [output for output in original_model.graph.output]
    initializers = [initializer for initializer in original_model.graph.initializer]
    value_info = [info for info in original_model.graph.value_info]

    new_nodes = []
    for idx, node in enumerate(model.graph.node):
        # Instrument each node and add to new nodes list
        new_node = instrument_node(node, model, idx, new_nodes)
        new_nodes.append(new_node)
        
    # Create a new graph with the instrumented nodes and copied components
    instrumented_graph = onnx.helper.make_graph(
        nodes=new_nodes,
        name=original_model.graph.name,
        inputs=inputs,
        outputs=outputs,
        initializer=initializers,
        value_info=value_info
    )

    return instrumented_graph

def separate_outputs(original_output_names, outputs):
    original_outputs = {}
    instrumented_outputs = {}

    for name, value in zip(session.get_outputs(), outputs):
        if name.name in original_output_names:
            original_outputs[name.name] = value
        else:
            instrumented_outputs[name.name] = value

    return original_outputs, instrumented_outputs

def create_state_transition_graph(instrumented_outputs):
    graph = {}
    for node_name, output in instrumented_outputs.items():
        # Assuming the node_name is structured as {original_node_name}_instrumented_{idx}
        original_node_name = '_'.join(node_name.split('_')[:-2])
        graph[original_node_name] = {'output': output}  # Store output data
        # You can add input data and other relevant information as well
    return graph

import hashlib

def hash_tensor_data(data):
    # Convert data to a byte representation and hash
    return hashlib.sha256(data.tobytes()).hexdigest()

def create_checkpoints(graph):
    checkpoints = []
    for node, data in graph.items():
        checkpoint = {
            'partition_id': node,
            'input_hashes': [],  # Fill with actual input hashes if available
            'output_hashes': [hash_tensor_data(data['output'])]
        }
        checkpoints.append(checkpoint)
    return checkpoints

def inference_instrumented_onnx_model(instrumented_onnx_model, input_tensor, original_output_names):
    '''
    Runs an inference on an instrumented onnx model, returning the main output of the onnx graph, along with checkpoints.
    
                Parameters:
                        instrumented_onnx_model (ONNXModel): The onnx model to instrument
                        input_tensor (tensor): The input tensor to run an inference on
    
                Returns:
                        output_tensor (tensor): The output tensor of the onnx model
                        checkpoints (List<Checkpoint>): The checkpoints of all the instrumented intermediate nodes
    '''
    # instrumented_onnx_model.eval()
    session = ort.InferenceSession(instrumented_onnx_model)
    input_name = session.get_inputs()[0].name
    output, *instrumented_outputs = session.run(None, {input_name: input_data})

    return output
```

```python
import onnx
from onnx import helper
import copy

# Function to infer the shapes and types of all tensors in the model
def infer_shapes_and_types(model):
    # Apply shape inference on the model
    inferred_model = onnx.shape_inference.infer_shapes(model)
    return inferred_model

def get_output_info(model, output_name):
    # Search in graph's value_info which now contains inferred shapes and types
    for value_info in model.graph.value_info:
        if value_info.name == output_name:
            return value_info.type
    # Search in graph's output
    for output in model.graph.output:
        if output.name == output_name:
            return output.type
    raise ValueError(f"Output type for {output_name} not found.")
    
def add_identity_nodes(model, graph):
    identity_nodes = []
    new_outputs = []

    for node in graph.node:
        for output in node.output:
            identity_output_name = f"{output}_instrumented"
            identity_nodes.append(
                onnx.helper.make_node(
                    'Identity',
                    inputs=[output],
                    outputs=[identity_output_name],
                    name=identity_output_name
                )
            )
            output_type = get_output_info(model, output)
            new_outputs.append(helper.make_value_info(identity_output_name, output_type))

    return identity_nodes, new_outputs

def instrument_onnx_graph(model):
    # Infer shapes and types
    model_with_inferred_shapes = infer_shapes_and_types(model)

    cloned_model = copy.deepcopy(model_with_inferred_shapes)
    identity_nodes, new_outputs = add_identity_nodes(cloned_model, cloned_model.graph)

    # Append the new nodes and outputs after completing the loop
    cloned_model.graph.node.extend(identity_nodes)
    cloned_model.graph.output.extend(new_outputs)

    # Check the modified model for correctness
    onnx.checker.check_model(cloned_model)

    return cloned_model
```

```python
import networkx as nx
import hashlib
import pickle
import base64

def hash_tensor_data(data):
    # Convert data to a byte representation and hash
    return hashlib.sha256(data.tobytes()).hexdigest()

def extract_identity_outputs(graph, instrumented_outputs):
    identity_output_map = {}

    # Counter to keep track of the position in instrumented_outputs
    output_idx = 0

    # Map each original node's output name to its corresponding identity output tensor
    for node in graph.node:
        for output in node.output:
            if output.endswith("_instrumented"):
                continue
            # Map the original output to the corresponding tensor from instrumented_outputs
            if output_idx < len(instrumented_outputs):
                identity_output_map[output] = instrumented_outputs[output_idx]
                output_idx += 1

    return identity_output_map

def construct_state_transition_graph(graph, identity_output_map):
    state_transition_graph = nx.DiGraph()

    # Add nodes to the graph
    for node in graph.node:
        if node.name.endswith('_instrumented'):
            continue
        state_transition_graph.add_node(node.name)

    # Iterate through each node to establish edges based on the node outputs
    for node in graph.node:
        for output in node.output:
            if output.endswith('_instrumented'):
                continue

            # Get the consumer nodes that take this output as an input
            consumer_nodes = [consumer for consumer in graph.node if output in consumer.input and not consumer.name.endswith('_instrumented')]
            
            # Add edges from the current node to the consumer nodes
            for consumer in consumer_nodes:
                if output in identity_output_map:
                    print(f'{node.name} - - - ({output}) - - - > {consumer.name}')
                    print(base64.b64encode(pickle.dumps(identity_output_map[output])).decode('utf-8'))
                    # Add an edge with the hash of the current node's Identity output data
                    state_transition_graph.add_edge(
                        node.name, consumer.name, 
                        hash=hash_tensor_data(identity_output_map[output])[0]
                    )

    return state_transition_graph
```

```python
import onnxruntime as ort
import numpy as np
import time
import random

def benchmark_model(model_path, test_inputs, runs=10000):
    # Load the ONNX model
    session = ort.InferenceSession(model_path)

    # Warm up the session
    _ = session.run(None, {"onnx::Gemm_0": test_inputs[0]})

    # Measure inference time
    start_time = time.time()
    for i in range(runs):
        outputs = session.run(None, {"onnx::Gemm_0": test_inputs[i % len(test_inputs)]})
    end_time = time.time()

    # Compute average time per inference
    avg_time_per_inference = (end_time - start_time) / runs
    return avg_time_per_inference

# Prepare a set of random inputs from the test dataset
num_inputs = len(test_dataset)  # Number of different inputs for benchmarking
test_inputs = [image.numpy().reshape(1, 28*28).astype(np.float32) for (image, label) in test_dataset]

# Benchmark the original model
avg_time_original = benchmark_model('model.onnx', test_inputs)
print(f"Average inference time for original model: {avg_time_original:.6f} seconds")

# Benchmark the instrumented model
avg_time_instrumented = benchmark_model('instrumented_model.onnx', test_inputs)
print(f"Average inference time for instrumented model: {avg_time_instrumented:.6f} seconds")

ratio = avg_time_instrumented / avg_time_original
print(f'Instrumented version is: {ratio:.2f} times slower than original')
```
